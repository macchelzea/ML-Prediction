```python
import pandas as pd
```

```python
df = pd.read_csv("Visadataset.csv")
df.head(3)
```

```python
df.shape
```

```python
# Converting the df to dictionary for the purpose of storing on mongodb
data = df.to_dict(orient='records')
```

```python
len(data)
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
# CONNECTION_URL2 = os.getenv('MONGODB_URL_KEY')
# CONNECTION_URL3=os.environ["MONGODB_URL_KEY"]
```

```python
# Now define the required variable for Mongodb
DB_NAME = 'mlops_db'
COLLECTION_NAME = 'travel'
CONNECTION_URL =  os.getenv('MONGODB_URL_KEY')
```

```python
import pymongo
```

```python
client = pymongo.MongoClient(CONNECTION_URL)
db = client[DB_NAME]
collection = db[COLLECTION_NAME]

#Now to ingest the data into Mongodb
records = collection.insert_many(data)
```

```python
# To load the ingested data back to the notebook
reload_records = collection.find()
```

```python
# Better still, the reloaded records can be inform of a df 
reload_df = pd.DataFrame(list(collection.find()))
reload_df.head(3)
```

```python
# _id is always added to the recodrd when ingesting into mongo db
# To remove this _id, simply do:
if "_id" in reload_df.columns.to_list():
    reload_df = reload_df.drop("_id", axis=1)
```
